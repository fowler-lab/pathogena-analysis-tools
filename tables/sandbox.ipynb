````python
import json 

def parse_variants(row):

    variant = row.variant
    is_null = False
    is_minor = False
    minor_variant = None
    minor_reads = None
    coverage = None
    frs = None

    a=(json.loads(row['vcf_evidence']))
    if 'COV_TOTAL' in a.keys():
        coverage = a['COV_TOTAL']
    if 'FRS' in a.keys():
        frs = a['FRS']

    if row.variant[-1]=='x':
        is_null = True
    elif ':' in row.variant:
        is_minor = True
        minor_variant, minor_reads = row.variant.split(':')
        if "ins" in variant or "del" in variant:
            variant = minor_variant.split('_')[0]+"_minorindel"
        else:
            variant = minor_variant[:-1]+'x'

    return pandas.Series([variant, is_null, is_minor, minor_variant, minor_reads, coverage, frs])

variants[['var','is_null','is_minor','minor_variant','minor_reads','coverage','FRS']] = variants.apply(parse_variants, axis=1)
````
